```python
from selenium import webdriver
import os
from selenium.webdriver.common.by import By
import time
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.support.ui import WebDriverWait
import pandas as pd
```

```python
#open the page

os.environ['PATH'] += r"C:/SeleniumDrivers"
driver = webdriver.Chrome()
driver.get("https://www.pathwaystoscience.org/programs.aspx?u=&d=GEN-STEM_STEM+Fields&d=MED-_Medical+%26+Life+Sciences+(All)&d=ENG-_Engineering+%26+Physical+Sciences+(All)&d=MAT-_Mathematics+%26+Computational+Sciences+(All)&d=EAR-_Earth%2C+Environmental+%26+Agriculture+Sciences+(All)&d=TEC-_Technology+(All)&submit=y&all=all")
time.sleep(10)
```

```python
#list will store links to the webpages of scholarships

lst = []
```

```python
# Create list with links to scholarships
def find_xpath(driver, xpath, timeout=100):
    # Wait for the visibility of elements using WebDriverWait
    WebDriverWait(driver, timeout).until(EC.visibility_of_all_elements_located((By.XPATH, xpath)))

    # Find the link elements
    link_elements = driver.find_elements(By.XPATH, xpath )

    # Iterate over the link elements and extract href
    for link in link_elements:
        href = link.get_attribute('href')
        if href:
            lst.append(href)
            print(link)
            
    return lst
```

```python
#iterates through links

xpath_1 = "//a[@href]"
find_xpath(driver, xpath = xpath_1)
```

```python
#makes sure all links are from the website 
filtered_lst = [link for link in lst if link.startswith("https://www.pathwaystoscience.org/programhub.aspx?sort=")]
```

```python
# using list comprehension to remove duplicated from list
res = []
[res.append(x) for x in filtered_lst if x not in res]
```
